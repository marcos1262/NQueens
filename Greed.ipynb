```python
import sys
from joblib import Parallel, delayed
from time import time

from NQueens import *

N = 10000
t = np.random.permutation(N)

# print(t)
# print_t(t)

# print(h(t))
# print(h2(t))
# print(h3(t))
```

```python
def successor(t, h_t, i):
    min = sys.maxsize
    best = (h_t, list(t))
    h_s = (0, [])
    
    for j in np.random.choice(range(N), int(N/10) if N > 50 and h_s[0] > 10 else N):
        h_s = best[0]
        s = best[1]
        
        h_s = (h_s[0] - individual_h2(s, j, i), h_s[1])
        s[i], s[j] = s[j], s[i]
        h_s = (h_s[0] + individual_h2(s, j, i), h_s[1])
        # h_s = h2(s)

        # better heuristic or less conflicts
        if h_s[0] < min or (h_s[0] == min and len(h_s[1]) < len(best[0][1])):
            min = h_s[0]
            best = (h_s, s)
        else:
            s[i], s[j] = s[j], s[i]

    return best


def bestSuccessor(t, h_t):
    min = sys.maxsize
    best = None

    successors = Parallel(n_jobs=-1)(delayed(successor)(t, h_t, i)
                                     for i in np.random.choice(h_t[1], int(10) if N > 50 else N))
    
    for s in successors:
        if s[0][0] < min or (s[0][0] == min and len(s[0][1]) < len(best[0][1])):
            min = s[0][0]
            best = s

    return h3(best[1]), best[1]


start = time()
h_t = h3(t)
print("Initial Heuristic:", h_t[0])
print("Best successor:", bestSuccessor(t, h_t)[0])
elapsed = time() - start
print("Elapsed (sec):", elapsed)
```

```python
def hill(initial):
    t = list(initial)
    h_t = h3(t)
    steps = 0

    while h_t[0] > 0:
        if steps % 10 == 0:
            print("h(t) at", steps, "steps:", h_t)
        steps += 1

        h_t1, t1 = bestSuccessor(t, h_t)

        if h_t1[0] < h_t[0] or (h_t1[0] == h_t[0] and h_t1 != h_t):
            h_t = h_t1
            t = t1
        else:
            # random movement
            i, j = np.random.choice(range(N), 2)
            h_t = (h_t[0] - individual_h2(t, i, j), h_t[1])
            t[i], t[j] = t[j], t[i]
            h_t = (h_t[0] + individual_h2(t, i, j), h_t[1])

    return t, steps


start = time()

solution, steps = hill(t)
print("Solution, steps:", solution, steps)

# print_t(solution)
print(h(solution))

elapsed = time() - start
print("Elapsed (sec):", elapsed)
```
